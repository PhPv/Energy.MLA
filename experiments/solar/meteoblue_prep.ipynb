```python
# ds = 'Meteoblue_ERA_ dataexport_20230215T052109.csv'
ds = 'Meteoblue_NEMS_dataexport_20230215T051617.csv'
```

```python
df = pd.read_csv(p/ds, sep=',', skiprows=9, parse_dates=['timestamp'])
```

```python
df = df.rename(mapper={'timestamp': 'dt'}, axis=1)
```

```python
df
```

```python
p = '/home/kpavel/PycharmProjects/energy/data/solar/station_vinc/train/fact.csv'
dft = pd.read_csv(p, sep=',', parse_dates=['Date\Time'])
dft = dft.sort_values('Date\Time').reset_index(drop=True)
dft.columns = ['dt', 'target']
dft
```

```python
dfj = df.merge(dft, on='dt', how='left')
```

```python
dfj['target'] = dfj.target.fillna(0)
dfj
```

```python
dfj.index = dfj['dt']
dfj = dfj[dfj.columns[1:]]
dfj
```

```python
dfj.to_csv('/home/kpavel/PycharmProjects/energy/data/solar/station_vinc/train/Meteoblue_NEMS_dataexport_20230215T051617_prep.csv', sep=',')
```
